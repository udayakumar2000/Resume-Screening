```python
data = pd.read_csv('pred_data.csv')
data = data.iloc[:,1:]

data.head()
```

```python
# useing PyResparser to extract information from the resume
extracted_job_info = pyresparser.ResumeParser(f"job.docx").get_extracted_data()['skills']
```

```python
extracted_job_info
```

```python
job_constraints =  extracted_job_info['skills']

job_constraints
```

```python
# create a vectorizer to convert the job constraints and candidate skills to vectors
vectorizer = TfidfVectorizer()

# convert the job constraints list to a string
job_constraints_str = ' '.join(job_constraints)

# convert the job constraints to a vector
job_constraints_vec = vectorizer.fit_transform([job_constraints_str.lower()])

# convert the candidate skills to vectors
candidate_skills_vecs = vectorizer.transform(data['Skills'])

# calculate the cosine similarity between the job constraints vector and each candidate's skills vector
cos_sim_scores = cosine_similarity(job_constraints_vec, candidate_skills_vecs).flatten()

# add the cosine similarity scores as a new column in the dataframe
data['cos_sim_score'] = cos_sim_scores

# sort the dataframe by the cosine similarity score in descending order
data = data.sort_values('cos_sim_score', ascending=False)
```
